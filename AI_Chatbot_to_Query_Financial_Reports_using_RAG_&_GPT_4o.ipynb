```python
vectorstore = FAISS.from_documents(documents, hf)
```

```python
def detect_report_type_via_llm(query: str) -> str:
    system_msg = "You're an financial assistant that determines if the question is about a quarterly or annual report."
    prompt = f"{system_msg}\n\nQuestion: {query}\nAnswer with just 'quarterly', 'annual', or 'unknown'."
    response = llm.predict(prompt)
    return response.strip().lower()
```

```python
from langchain.schema import BaseRetriever, Document
from langchain.vectorstores.base import VectorStoreRetriever
from typing import List
from pydantic import Field

class MetadataAwareRetriever(BaseRetriever):
    base_retriever: VectorStoreRetriever = Field()

    def get_relevant_documents(self, query: str) -> List[Document]:
        report_filter = detect_report_type_via_llm(query)
        all_docs = self.base_retriever.get_relevant_documents(query)
        if report_filter== "quarterly" or report_filter == "annual":
            filtered_docs = [
                doc for doc in all_docs if doc.metadata.get("report_type") == report_filter
            ]
            return filtered_docs if filtered_docs else all_docs
        return all_docs
```

```python
base = vectorstore.as_retriever(search_kwargs={"k": 5})
metadata_retriever = MetadataAwareRetriever(base_retriever=base)
```

```python
from langchain.prompts import PromptTemplate

custom_prompt_template = """
You are a highly intelligent Financial AI Assistant. Your job is to help users understand corporate financial reports.
Use the provided context to give detailed, fact-based answers. If asked to explain financial tables like balance sheets, income statements, or cash flow statements, break them down clearly.

Guidelines:
- Always cite and reference the actual report data
- DO NOT make up numbers or fabricate content
- Maintain context from previous questions in the conversation
- Use simple explanations if the user asks for clarity
- Do not include anything related to "but the context provided specifically" in the answer

Chat History:
{chat_history}

User Question:
{question}

Relevant Context:
{context}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["chat_history", "question", "context"],
    template=custom_prompt_template
)
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

question_gen_prompt = PromptTemplate.from_template("""
Given the following conversation history and a follow-up question, rephrase the follow-up question to be a standalone question.

Chat History:
{chat_history}

Follow-up question:
{question}

Standalone question:
""")

question_generator = LLMChain(
    llm=llm,
    prompt=question_gen_prompt
)
```

```python
from langchain.chains.question_answering import load_qa_chain

combine_docs_chain = load_qa_chain(
    llm=llm,
    chain_type="stuff",
    prompt=prompt,
)
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain

memory = ConversationBufferMemory(
    memory_key="chat_history",
    input_key="question",
    output_key="answer",
    return_messages=True
)

qa_chain = ConversationalRetrievalChain(
    retriever=metadata_retriever,
    combine_docs_chain=combine_docs_chain,
    question_generator=question_generator,
    memory=memory,
    return_source_documents=True,
    output_key="answer"
)
```
